```python
testPics = []

for subDir in os.listdir(testDir):
    subDirFullPath = os.path.join(testDir, subDir)
    if(os.path.isfile(subDirFullPath)):
        continue
    print("'" + subDirFullPath + "' contains " + str(len(os.listdir(subDirFullPath))) + " contents.")
    for f in os.listdir(subDirFullPath):
        if(os.path.isdir(f)):
            continue
        testPics.append(os.path.join(subDirFullPath, f))

# testPics
```

```python
counter = 0
right = 0
wrong = []
for pic in testPics:
    img = image.load_img(pic , target_size=(100,100))
    img = image.img_to_array(img)
    img = img.reshape((1,) + img.shape)
    img_class=model.predict(img, verbose=1)
    prediction = img_class[0]
    # print(prediction)
    counter = counter + 1
    for i in range(len(classname)):
        if(prediction[i] == 1):
            if("\\" + classname[i] + "\\" in pic):
                right = right + 1
            # print("'" + pic + "' => '" + classname[i] + "'")
            else:
                wrong.append("'" + pic + "' => '" + classname[i])
                print("'" + pic + "' => '" + classname[i] + "' ---> Guessed wrong")
```

```python
print("Total Data = " + str(counter))
print("Right pridiction = " + str(right))
print("Right pridiction % = " + str(100 * right/counter))
print("Wrong pridiction = " + str(len(wrong)))
print("Wrong pridiction % = " + str(100 * len(wrong)/counter))
```
